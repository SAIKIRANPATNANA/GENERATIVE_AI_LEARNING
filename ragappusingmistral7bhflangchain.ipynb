## Defining nlp methods of text preprocessing

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

import nltk
nltk.download('stopwords')
nltk.download('punkt')

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if word.lower() not in stop_words]
    return ' '.join(filtered_text)

def remove_citations(text):
    citation_pattern = r'\[\d+\]'
    return re.sub(citation_pattern, '', text)

def remove_stopwords_and_citations(text):
    text_without_stopwords = remove_stopwords(text)
    text_without_citations = remove_citations(text_without_stopwords)
    return text_without_citations
```

## Defining prompt template to prompt the model

```python
from langchain.prompts import ChatPromptTemplate
template="""You are an assistant for question-answering tasks.
Use the following pieces of retrieved context to answer the question.
If you don't know the answer, just say that you don't know.
keep the answer concise.
Question: {question}
Context: {context}
Answer:
"""
```

## Integrating with hugging face and defining the mistral 7b model

```python
prompt = ChatPromptTemplate.from_template(template)
```

```python
from langchain import HuggingFaceHub
```

```python
from google.colab import userdata
huggingfacehub_api_token=userdata.get('HUGGING_API_KEY')
```

```python

model = HuggingFaceHub(
    huggingfacehub_api_token=huggingfacehub_api_token,
    repo_id="mistralai/Mistral-7B-Instruct-v0.1",
    model_kwargs={"temperature":1, "max_length":100}
)
```

```python

from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser
```

```python
output_parser=StrOutputParser()
```

```python
retriever=vector_db.as_retriever()
```

## defining the rag chain by integrating prompt, context, query

```python
rag_chain = (
    {"context": retriever,  "question": RunnablePassthrough()}
    | prompt
    | model
    | output_parser
)
```
